# Example

```python
startgamesetup(None, sheets=[["F"]], variants=["use version 2.4 rules","use house rules"], style="tropical", dotsperhex=300)

x = 3
y = 3
for s in [
    "infantry", "armor", "artillery", "reconnaissance", "antiarmor", "airdefense", "supply", "transportation", "ammunition", "fuel", "ordnance",
    "headquarters", "light", "medium", "heavy", "missile", "gun", "multiplerocket", "radar", "motorized", "wheeled", "limitedwheeled",
    "locomotive", "railcar", "truck", "barge", "junk", "building", "tower", "hangar", "fixedwing", "rotarywing", "attack", "bomber", "cargo", "fighter", "fighterbomber", "patrol", "utility"
    ]:
    hexcode = "F-%02d%02d" % (x, y)
    groundunit(s, hexcode, symbols=s, color="lightgreen")
    y += 1
    if y == 12:
        x += 1
        y = 3

x = 9
y = 3
for c in [ "white", "natoblue", "natored", "natogreen", "natoyellow", "green", "lightgreen", "tan", "darktan", "sand" ]:
    hexcode = "F-%02d%02d" % (x, y)
    groundunit(c, hexcode, "infantry", color=c)
    y += 1
    if y == 8:
        x += 1
        y = 3

x = 9
y = 9
hexcode = "F-%02d%02d" % (x, y)
groundunit("1/1", hexcode, "infantry", stack="1/1", color="lightgreen")
y += 1
hexcode = "F-%02d%02d" % (x, y)
groundunit("1/2", hexcode, "infantry", stack="1/2", color="lightgreen")
groundunit("2/2", hexcode, "infantry", stack="2/2", color="lightgreen")
x += 1
y -= 1
hexcode = "F-%02d%02d" % (x, y)
groundunit("1/3", hexcode, "infantry", stack="1/3", color="lightgreen")
groundunit("2/3", hexcode, "infantry", stack="2/3", color="lightgreen")
groundunit("3/3", hexcode, "infantry", stack="3/3", color="lightgreen")
y += 1
hexcode = "F-%02d%02d" % (x, y)
groundunit("1/4", hexcode, "infantry", stack="1/4", color="lightgreen")
groundunit("2/4", hexcode, "infantry", stack="2/4", color="lightgreen")
groundunit("3/4", hexcode, "infantry", stack="3/4", color="lightgreen")
groundunit("4/4", hexcode, "infantry", stack="4/4", color="lightgreen")


x = 12
y = 3
for s in [ 
    "infantry/armor", "infantry/armor/gun", "infantry/armor/wheeled", "infantry/light", "armor/medium/gun", "artillery/gun", "artillery/armor/gun", "artillery/multiplerocket/limitedwheeled", 
    "airdefense/gun/light", "airdefense/missile", "airdefense/radar", "airdefense/gun/armor/medium", "airdefense/missile/wheeled", "airdefense/missile/armor", "airdefense/headquarters", "fixedwing/fighter", "rotarywing/cargo",
    ]:
    hexcode = "F-%02d%02d" % (x, y)
    groundunit(s, hexcode, symbols=s, color="lightgreen")
    y += 1
    if y == 12:
        x += 1
        y = 3

endgamesetup()
drawmap(zoom=False, xmin=2, xmax=16, ymin=13, ymax=23.5)
```
